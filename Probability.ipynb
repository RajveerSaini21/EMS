```python
import numpy as np 
import matplotlib.pyplot as plt 
import random 



#Toss a coin 10000 times ---> Toss a coin 10000 times and compute probability of Heads and Tails
def coin_tosses(n = 10000):
    heads = 0 
    for _ in range(n):
        if random.choice(['Heads' , 'Tails']) == 'Heads' :
            heads += 1 
    
    tails = n - heads
    print(f"Coin Toss - Heads: {heads/n:.4f}, Tails: {tails/n:.4f}")


#Roll two dice 10000 times and calculate probability of getting sum = 7
def dice_roll(n = 10000):
    count = 0 
    for _ in range(n):
        if random.randint(1,6) + random.randint(1,6) == 7 :
            count += 1 
    print(f"Dice sum = 7\nProbability : {count/n:.4f}")


# At least one 6 in 10 ---> Probability of getting at least one 6 in 10 dice rolls
def atleast_one_six(n = 10000):
    count = 0 ; 
    for _ in range(n):
        if 6 in [random.randint(1,6) for _ in range(10)]:
            count += 1        
    print(f"At least one 6 in 10 rolls: {count/n:.4f}")
    

# Conditional Probability: P(Red | Blue before)
def conditional_probability(k=1000):
    bag = ['Red']*5 + ['Green']*7 + ['Blue']*8
    draws = [random.choice(bag) for _ in range(k)]
    
    blue_before = 0 
    red_after_blue = 0 
    
    for i in range(1 , k):
        if(draws[i-1] == 'Blue'):
            blue_before += 1 ; 
            if(draws[i] == 'Red'):
                red_after_blue += 1 
    
    # Calculate conditional probability
    if blue_before != 0:
        probability = red_after_blue / blue_before
    else:
        probability = 0 
    
    print(f"P(Red | Previous was Blue): {probability:.4f}")



# Discrete Random Variable ---> Simulate a discrete random variable and calculate statistics
def discrete_random_values():
    values = [1,2,3]
    probs = [0.25,0.35,0.4]
    
    samples = np.random.choice(values, size=1000, p=probs)
    mean = np.mean(samples)
    variance = np.var(samples)
    std_dev = np.std(samples)
    
    print(f"Mean: {mean:.4f}, Variance: {variance:.4f}, Std Dev: {std_dev:.4f}")


# Plot Exponential Distribution
def plot_exponential():
    data = np.random.exponential(scale=5, size=2000)
    plt.hist(data, bins=50, density=True, color="skyblue", alpha=0.7)
    plt.title("Exponential Distribution (mean=5)")
    plt.xlabel("Values")
    plt.ylabel("Density")
    plt.show()


#Central Limit Theorem simulation
def central_limit_theorem_demo():
    population = np.random.uniform(0,1,10000)
    sample_means=[]
    
    # Take 1000 samples of size 30 and compute their means
    for _ in range(1000):
        sample = np.random.choice(population, size=30)
        sample_means.append(np.mean(sample))
        
    plt.hist(sample_means, bins=30, color='orange', alpha=0.7)
    plt.title("Central Limit Theorem")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.show()
    

if __name__ == '__main__':
    coin_tosses()
    dice_roll()
    atleast_one_six()
    conditional_probability()
    discrete_random_values()
    
    print("\nShowing Exponential Distribution Plot...")
    plot_exponential()
    
    print("Showing Central Limit Theorem Plot...")
    central_limit_theorem_demo()

    


    
    
    
```
